## 1. Tangente und Normale an einem Punkt

Die **Tangente** an f im Punkt (x₀, f(x₀)) hat die Gleichung:

$$t(x) = f'(x_0) \cdot (x - x_0) + f(x_0)$$

Die **Normale** steht senkrecht auf der Tangente. Ihre Steigung ist $-\frac{1}{f'(x_0)}$ (falls $f'(x_0) \neq 0$):

$$n(x) = -\frac{1}{f'(x_0)} \cdot (x - x_0) + f(x_0)$$

Verschiebe x₀ und beobachte, wie sich Tangente und Normale ändern.

```python
def tangente_normale(x0=1.0):
    """Visualisiert Tangente und Normale an f(x) = x³ - 3x."""
    f = lambda x: x**3 - 3*x
    f_prime = lambda x: 3*x**2 - 3

    x = np.linspace(-3, 3, 500)
    y0 = f(x0)
    m = f_prime(x0)

    # Tangente
    t = m * (x - x0) + y0

    fig, ax = plt.subplots()
    ax.plot(x, f(x), 'b-', linewidth=2.5, label=r'$f(x) = x^3 - 3x$')
    ax.plot(x, t, 'r--', linewidth=1.8,
            label=rf'Tangente: $t(x) = {m:.2f}(x - {x0:.1f}) + ({y0:.2f})$')

    # Normale (nur wenn Steigung nicht 0)
    if abs(m) > 1e-6:
        m_n = -1 / m
        n = m_n * (x - x0) + y0
        ax.plot(x, n, 'g-.', linewidth=1.8,
                label=rf'Normale ($m_n = {m_n:.2f}$)')

        # Rechter Winkel andeuten
        winkel_laenge = 0.25
        dx_t = winkel_laenge / np.sqrt(1 + m**2)
        dy_t = m * dx_t
        dx_n = winkel_laenge / np.sqrt(1 + m_n**2)
        dy_n = m_n * dx_n
        eck_x = [x0 + dx_t, x0 + dx_t + dx_n, x0 + dx_n]
        eck_y = [y0 + dy_t, y0 + dy_t + dy_n, y0 + dy_n]
        ax.plot(eck_x, eck_y, 'k-', linewidth=1, alpha=0.6)
    else:
        # Senkrechte Normale
        ax.axvline(x=x0, color='g', linestyle='-.', linewidth=1.8,
                   label=r'Normale (senkrecht)')

    ax.plot(x0, y0, 'ro', markersize=10, zorder=5)
    ax.set_xlim(-3, 3)
    ax.set_ylim(-8, 8)
    ax.axhline(y=0, color='k', linewidth=0.5)
    ax.axvline(x=0, color='k', linewidth=0.5)
    ax.grid(True, alpha=0.3)
    ax.legend(fontsize=11, loc='upper left', framealpha=0.9)
    ax.set_title(rf'Tangente & Normale bei $x_0 = {x0:.1f}$, Steigung $m = {m:.2f}$',
                 fontsize=14)
    plt.tight_layout()
    plt.show()

interact(
    tangente_normale,
    x0=FloatSlider(value=1.0, min=-2.5, max=2.5, step=0.1,
                   description='x\u2080:', style={'description_width': '30px'},
                   layout={'width': '400px'})
);
```

**Beobachte:**
- Bei x₀ = 1 und x₀ = -1 ist die Tangente waagrecht (m = 0) — das sind die **Extremstellen**!
- Die Normale steht dort senkrecht.
- Je steiler die Tangente, desto flacher die Normale (und umgekehrt).

## 2. Gemeinsame Tangenten zweier Funktionen

Zwei Funktionen f und g haben im Punkt x₀ eine **gemeinsame Tangente**, wenn:
1. f(x₀) = g(x₀) (gleicher Funktionswert)
2. f'(x₀) = g'(x₀) (gleiche Steigung)

Man sagt: Die Graphen **berühren** sich in x₀.

Verschiebe den Slider und finde den Berührpunkt!

```python
def gemeinsame_tangente(x0=2.0):
    """Zeigt f(x) = x² und g(x) = -x² + 4x mit Tangenten bei x₀."""
    f = lambda x: x**2
    f_prime = lambda x: 2*x
    g = lambda x: -x**2 + 4*x
    g_prime = lambda x: -2*x + 4

    x = np.linspace(-1, 5, 500)

    y_f = f(x0)
    y_g = g(x0)
    m_f = f_prime(x0)
    m_g = g_prime(x0)

    t_f = m_f * (x - x0) + y_f
    t_g = m_g * (x - x0) + y_g

    fig, ax = plt.subplots()
    ax.plot(x, f(x), 'b-', linewidth=2.5, label=r'$f(x) = x^2$')
    ax.plot(x, g(x), 'purple', linewidth=2.5, label=r'$g(x) = -x^2 + 4x$')

    ax.plot(x, t_f, 'r--', linewidth=1.5, alpha=0.8,
            label=rf'Tangente an f ($m = {m_f:.1f}$)')
    ax.plot(x, t_g, 'orange', linestyle='--', linewidth=1.5, alpha=0.8,
            label=rf'Tangente an g ($m = {m_g:.1f}$)')

    ax.plot(x0, y_f, 'ro', markersize=9, zorder=5)
    ax.plot(x0, y_g, 'o', color='purple', markersize=9, zorder=5)

    # Berührpunkt-Check
    wert_gleich = abs(y_f - y_g) < 0.05
    steigung_gleich = abs(m_f - m_g) < 0.05
    if wert_gleich and steigung_gleich:
        ax.set_title(r'$\checkmark$ Berührpunkt gefunden! Gleicher Wert UND gleiche Steigung!',
                     fontsize=14, color='green', fontweight='bold')
    else:
        hinweise = []
        if not wert_gleich:
            hinweise.append(rf'$f({x0:.1f}) = {y_f:.2f} \neq {y_g:.2f} = g({x0:.1f})$')
        if not steigung_gleich:
            hinweise.append(rf"$f'({x0:.1f}) = {m_f:.1f} \neq {m_g:.1f} = g'({x0:.1f})$")
        ax.set_title('  |  '.join(hinweise), fontsize=12, color='red')

    ax.set_xlim(-1, 5)
    ax.set_ylim(-2, 10)
    ax.axhline(y=0, color='k', linewidth=0.5)
    ax.axvline(x=0, color='k', linewidth=0.5)
    ax.grid(True, alpha=0.3)
    ax.legend(fontsize=11, loc='upper left', framealpha=0.9)
    plt.tight_layout()
    plt.show()

interact(
    gemeinsame_tangente,
    x0=FloatSlider(value=0.5, min=0.0, max=4.0, step=0.1,
                   description='x\u2080:', style={'description_width': '30px'},
                   layout={'width': '400px'})
);
```

**Aufgabe:** Finde den Berührpunkt! (Tipp: Er liegt bei x₀ = 2.)

Rechnerisch: f(x) = g(x) und f'(x) = g'(x) gleichzeitig lösen:
- x² = -x² + 4x → 2x² - 4x = 0 → x(x - 2) = 0 → x = 0 oder x = 2
- 2x = -2x + 4 → 4x = 4 → x = 2

Nur bei x = 2 sind **beide** Bedingungen erfüllt!

## 2b. Tangente durch einen externen Punkt

Manchmal liegt der Punkt P **nicht** auf dem Graphen von f. Trotzdem kann man Tangenten an f finden, die durch P verlaufen — oft sogar **zwei**!

**Vorgehen:**
1. Allgemeine Tangente bei x₀ aufstellen: $t(x) = f'(x_0)(x - x_0) + f(x_0)$
2. Fordern, dass P auf der Tangente liegt: $t(x_P) = y_P$
3. Nach x₀ auflösen → ergibt die Berührstelle(n)

Verschiebe den Punkt P und beobachte, wie sich die Tangenten ändern!

```python
def tangente_externer_punkt(px=3.0, py=4.0):
    """Tangenten an f(x) = x² + 1 durch einen externen Punkt P."""
    f = lambda x: x**2 + 1
    f_prime = lambda x: 2*x

    x = np.linspace(-3, 5, 500)

    fig, ax = plt.subplots()
    ax.plot(x, f(x), 'b-', linewidth=2.5, label=r'$f(x) = x^2 + 1$')

    # Berührstellen finden: P liegt auf t(x) = f'(x0)(x - x0) + f(x0)
    # → py = 2*x0*(px - x0) + x0² + 1
    # → py = 2*x0*px - 2*x0² + x0² + 1
    # → py = 2*x0*px - x0² + 1
    # → x0² - 2*px*x0 + (py - 1) = 0
    diskr = px**2 - (py - 1)

    tangenten_info = []
    if diskr > 1e-6:
        x0_1 = px - np.sqrt(diskr)
        x0_2 = px + np.sqrt(diskr)
        tangenten_info = [x0_1, x0_2]
    elif abs(diskr) <= 1e-6:
        tangenten_info = [px]

    farben = ['r', 'orange']
    for i, x0 in enumerate(tangenten_info):
        m = f_prime(x0)
        t = m * (x - x0) + f(x0)
        ax.plot(x, t, '--', color=farben[i], linewidth=1.8,
                label=rf'Tangente bei $x_0 = {x0:.2f}$ ($m = {m:.2f}$)')
        ax.plot(x0, f(x0), 'o', color=farben[i], markersize=8, zorder=5)

    # Punkt P
    ax.plot(px, py, 'k*', markersize=15, zorder=6, label=rf'$P({px:.1f} \mid {py:.1f})$')

    # Status
    if diskr > 1e-6:
        ax.set_title(rf'Diskriminante $D = {diskr:.2f} > 0$ → zwei Tangenten',
                     fontsize=13, color='green')
    elif abs(diskr) <= 1e-6:
        ax.set_title(rf'$D = {diskr:.2f} \approx 0$ → genau eine Tangente (P liegt auf f)',
                     fontsize=13, color='orange')
    else:
        ax.set_title(rf'$D = {diskr:.2f} < 0$ → keine Tangente durch P möglich',
                     fontsize=13, color='red')

    ax.set_xlim(-3, 5)
    ax.set_ylim(-2, 15)
    ax.axhline(y=0, color='k', linewidth=0.5)
    ax.axvline(x=0, color='k', linewidth=0.5)
    ax.grid(True, alpha=0.3)
    ax.legend(fontsize=10, loc='upper left', framealpha=0.9)
    plt.tight_layout()
    plt.show()

interact(
    tangente_externer_punkt,
    px=FloatSlider(value=3.0, min=-2.0, max=4.5, step=0.1,
                   description='P x:', style={'description_width': '40px'},
                   layout={'width': '400px'}),
    py=FloatSlider(value=4.0, min=-1.0, max=12.0, step=0.1,
                   description='P y:', style={'description_width': '40px'},
                   layout={'width': '400px'}),
);
```

**Probiere aus:**
- **P(3|4):** Zwei Tangenten — der Standardfall aus Aufgabe A4
- **P nach oben schieben (z.B. P(0|1)):** P liegt auf dem Graphen → genau eine Tangente
- **P weit nach oben (z.B. P(0|12)):** P liegt "innerhalb" der Parabel → keine Tangente möglich (D < 0)

**Merke:** Die Anzahl der Tangenten hängt von der **Diskriminante** der quadratischen Gleichung ab:
- D > 0: zwei Tangenten
- D = 0: genau eine Tangente (P liegt auf dem Graphen)
- D < 0: keine Tangente

```python
scharen = {
    'f_t(x) = x³ − 3t²x': {
        'f': lambda x, t: x**3 - 3*t**2 * x,
        'label': r'$f_t(x) = x^3 - 3t^2 x$',
        'param': 't',
        'p_range': (0.5, 3.0, 0.1),
        'p_default': 1.0,
        'x_range': (-4, 4),
        'y_range': (-15, 15),
    },
    'f_a(x) = a·x² − 2x': {
        'f': lambda x, a: a * x**2 - 2*x,
        'label': r'$f_a(x) = a \cdot x^2 - 2x$',
        'param': 'a',
        'p_range': (0.2, 3.0, 0.2),
        'p_default': 1.0,
        'x_range': (-3, 5),
        'y_range': (-5, 10),
    },
    'f_a(x) = x·e^(−ax)': {
        'f': lambda x, a: x * np.exp(-a * x),
        'label': r'$f_a(x) = x \cdot e^{-ax}$',
        'param': 'a',
        'p_range': (0.2, 3.0, 0.2),
        'p_default': 1.0,
        'x_range': (-1, 6),
        'y_range': (-0.5, 1.5),
    },
}

def funktionsschar(wahl='f_t(x) = x³ − 3t²x', param_wert=1.0):
    info = scharen[wahl]
    f = info['f']
    x = np.linspace(*info['x_range'], 500)

    fig, ax = plt.subplots()

    # Aktuelle Kurve fett zeichnen
    ax.plot(x, f(x, param_wert), 'b-', linewidth=2.5,
            label=info['label'].replace(info['param'], f'{param_wert:.1f}'))

    # Mehrere Schar-Mitglieder dünn im Hintergrund
    p_min, p_max, p_step = info['p_range']
    for p in np.arange(p_min, p_max + p_step/2, p_step * 2):
        if abs(p - param_wert) > 0.05:
            ax.plot(x, f(x, p), '-', color='lightblue', linewidth=0.8, alpha=0.6)

    ax.set_xlim(*info['x_range'])
    ax.set_ylim(*info['y_range'])
    ax.axhline(y=0, color='k', linewidth=0.5)
    ax.axvline(x=0, color='k', linewidth=0.5)
    ax.grid(True, alpha=0.3)
    ax.legend(fontsize=13, loc='upper left', framealpha=0.9)
    ax.set_title(f'Funktionsschar mit {info["param"]} = {param_wert:.1f}', fontsize=14)
    plt.tight_layout()
    plt.show()

interact(
    funktionsschar,
    wahl=Dropdown(options=list(scharen.keys()), value='f_t(x) = x³ − 3t²x',
                  description='Schar:'),
    param_wert=FloatSlider(value=1.0, min=0.2, max=3.0, step=0.1,
                           description='Parameter:',
                           style={'description_width': '80px'},
                           layout={'width': '400px'})
);
```

**Beobachte:**
- Bei f_t(x) = x³ - 3t²x: Je größer t, desto weiter auseinander liegen die Extrempunkte.
- Bei f_a(x) = a·x² - 2x: Der Parameter a steuert die "Öffnung" der Parabel.
- Bei f_a(x) = x·e^(-ax): Je größer a, desto schneller fällt die Kurve ab und desto näher liegt das Maximum bei x = 0.

## 4. Ortskurve der Extrempunkte

Die **Ortskurve** verbindet alle Extrempunkte einer Funktionsschar, wenn man den Parameter variiert.

**Beispiel:** f_t(x) = x³ - 3t²x
- f_t'(x) = 3x² - 3t² = 0 → x = ±t
- Hochpunkt: H(-t | 2t³), Tiefpunkt: T(t | -2t³)
- Ortskurve der Hochpunkte: x = -t, y = 2t³ → t = -x → **y = -2x³**

```python
def ortskurve_explorer(t_aktuell=1.5):
    """Zeigt f_t(x) = x³ - 3t²x mit Extrempunkten und Ortskurve."""
    x = np.linspace(-4, 4, 500)

    fig, ax = plt.subplots()

    # Schar im Hintergrund
    for t in np.arange(0.5, 3.1, 0.25):
        y = x**3 - 3*t**2 * x
        ax.plot(x, y, '-', color='lightblue', linewidth=0.6, alpha=0.5)

    # Aktuelle Kurve
    y_aktuell = x**3 - 3*t_aktuell**2 * x
    ax.plot(x, y_aktuell, 'b-', linewidth=2.5,
            label=rf'$f_t(x) = x^3 - 3 \cdot {t_aktuell:.1f}^2 \cdot x$')

    # Extrempunkte der aktuellen Kurve
    x_hp = -t_aktuell
    y_hp = x_hp**3 - 3*t_aktuell**2 * x_hp  # = 2t³
    x_tp = t_aktuell
    y_tp = x_tp**3 - 3*t_aktuell**2 * x_tp  # = -2t³

    ax.plot(x_hp, y_hp, 'go', markersize=12, zorder=5, label=f'HP ({x_hp:.1f} | {y_hp:.1f})')
    ax.plot(x_tp, y_tp, 'rs', markersize=12, zorder=5, label=f'TP ({x_tp:.1f} | {y_tp:.1f})')

    # Ortskurve: y = -2x³ (Hochpunkte) und y = -2x³ (Tiefpunkte — gleiche Kurve!)
    x_ort = np.linspace(-3.5, 3.5, 300)
    y_ort = -2 * x_ort**3
    ax.plot(x_ort, y_ort, 'k--', linewidth=2, alpha=0.7, label=r'Ortskurve: $y = -2x^3$')

    # Alle Extrempunkte als Punkte auf der Ortskurve
    for t in np.arange(0.5, 3.1, 0.25):
        ax.plot(-t, 2*t**3, 'g.', markersize=5, alpha=0.6)
        ax.plot(t, -2*t**3, 'r.', markersize=5, alpha=0.6)

    ax.set_xlim(-4, 4)
    ax.set_ylim(-20, 20)
    ax.axhline(y=0, color='k', linewidth=0.5)
    ax.axvline(x=0, color='k', linewidth=0.5)
    ax.grid(True, alpha=0.3)
    ax.legend(fontsize=11, loc='upper left', framealpha=0.9)
    ax.set_title(rf'Funktionsschar $f_t(x) = x^3 - 3t^2 x$ mit Ortskurve', fontsize=14)
    plt.tight_layout()
    plt.show()

interact(
    ortskurve_explorer,
    t_aktuell=FloatSlider(value=1.5, min=0.5, max=3.0, step=0.1,
                          description='t:', style={'description_width': '30px'},
                          layout={'width': '400px'})
);
```

**Beobachte:**
- Die grünen Hochpunkte und roten Tiefpunkte wandern entlang der **Ortskurve** y = -2x³.
- Die Ortskurve ist selbst wieder eine Funktion — sie beschreibt den "Pfad" der Extrempunkte.

**Rezept: Ortskurve bestimmen**
1. Extrempunkte in Abhängigkeit vom Parameter berechnen: x_E(t), y_E(t)
2. Aus x_E(t) den Parameter t isolieren: t = ...
3. In y_E(t) einsetzen → y als Funktion von x → das ist die Ortskurve!

## 4b. Ortskurve der e-Funktion-Schar

**Beispiel:** $f_a(x) = x \cdot e^{-ax}$ mit $a > 0$

- $f_a'(x) = e^{-ax}(1 - ax) = 0 \Rightarrow x = \tfrac{1}{a}$
- Maximum bei $H\!\left(\tfrac{1}{a} \;\middle|\; \tfrac{1}{a \cdot e}\right)$
- Ortskurve: $x = \tfrac{1}{a} \Rightarrow a = \tfrac{1}{x}$, einsetzen: $y = x \cdot e^{-1/x \cdot x} = x \cdot e^{-1} = \tfrac{x}{e}$
- **Ortskurve: $y = \tfrac{x}{e}$** — eine Gerade!

```python
def ortskurve_efunktion(a_aktuell=1.0):
    """Zeigt f_a(x) = x·e^(-ax) mit Hochpunkt und Ortskurve y = x/e."""
    x = np.linspace(-0.5, 7, 500)

    fig, ax = plt.subplots()

    # Schar im Hintergrund
    for a in np.arange(0.3, 3.1, 0.3):
        y = x * np.exp(-a * x)
        ax.plot(x, y, '-', color='lightblue', linewidth=0.6, alpha=0.5)

    # Aktuelle Kurve
    y_aktuell = x * np.exp(-a_aktuell * x)
    ax.plot(x, y_aktuell, 'b-', linewidth=2.5,
            label=rf'$f_a(x) = x \cdot e^{{-{a_aktuell:.1f}\,x}}$')

    # Hochpunkt
    x_hp = 1 / a_aktuell
    y_hp = x_hp * np.exp(-a_aktuell * x_hp)  # = 1/(a·e)
    ax.plot(x_hp, y_hp, 'go', markersize=12, zorder=5,
            label=rf'HP $\left(\frac{{1}}{{{a_aktuell:.1f}}},\; \frac{{1}}{{{a_aktuell:.1f} \cdot e}}\right) = ({x_hp:.2f} \mid {y_hp:.3f})$')

    # Ortskurve: y = x/e
    x_ort = np.linspace(0, 5, 300)
    y_ort = x_ort / np.e
    ax.plot(x_ort, y_ort, 'k--', linewidth=2, alpha=0.7,
            label=r'Ortskurve: $y = \frac{x}{e}$')

    # Alle Hochpunkte auf der Ortskurve
    for a in np.arange(0.3, 3.1, 0.3):
        xh = 1 / a
        yh = 1 / (a * np.e)
        ax.plot(xh, yh, 'g.', markersize=6, alpha=0.6)

    ax.set_xlim(-0.5, 7)
    ax.set_ylim(-0.1, 1.5)
    ax.axhline(y=0, color='k', linewidth=0.5)
    ax.axvline(x=0, color='k', linewidth=0.5)
    ax.grid(True, alpha=0.3)
    ax.legend(fontsize=10, loc='upper right', framealpha=0.9)
    ax.set_title(rf'$f_a(x) = x \cdot e^{{-ax}}$ — Ortskurve der Hochpunkte', fontsize=14)
    plt.tight_layout()
    plt.show()

interact(
    ortskurve_efunktion,
    a_aktuell=FloatSlider(value=1.0, min=0.3, max=3.0, step=0.1,
                          description='a:', style={'description_width': '30px'},
                          layout={'width': '400px'})
);
```

**Beobachte:**
- Die Hochpunkte liegen alle auf der **Geraden** $y = \frac{x}{e}$ — überraschend einfach!
- Je größer a, desto näher rückt das Maximum an den Ursprung.
- Anders als bei der kubischen Schar ist die Ortskurve hier linear.

```python
from ipywidgets import Text

_x = sp.Symbol('x')

def _parse_formel(formel_str):
    """Parst einen Formelstring zu einem sympy-Ausdruck."""
    formel_normiert = formel_str.strip().replace('^', '**')
    return sp.sympify(
        formel_normiert,
        locals={
            'x': _x, 'e': sp.E,
            'ln': sp.log, 'log': sp.log, 'exp': sp.exp, 'sqrt': sp.sqrt,
            'sin': sp.sin, 'cos': sp.cos, 'tan': sp.tan,
        },
    )

def tangente_berechnen(f_sym, x0_wert):
    """Berechnet Tangenten- und Normalengleichung symbolisch."""
    f_prime_sym = sp.diff(f_sym, _x)
    y0 = f_sym.subs(_x, x0_wert)
    m = f_prime_sym.subs(_x, x0_wert)
    tangente = sp.expand(m * (_x - x0_wert) + y0)
    if m != 0:
        normale = sp.expand(-1/m * (_x - x0_wert) + y0)
    else:
        normale = None
    return tangente, normale, m, y0

def sympy_tangente(formel_str='x**3 - 3*x', x0=1.0):
    try:
        f_sym = _parse_formel(formel_str)
        x0_rational = sp.Rational(x0).limit_denominator(100)

        tangente, normale, m, y0 = tangente_berechnen(f_sym, x0_rational)

        display(Markdown(f"### Ergebnis"))
        display(Markdown(f"**f(x) =** ${sp.latex(f_sym)}$"))
        display(Markdown(f"**f'(x) =** ${sp.latex(sp.diff(f_sym, _x))}$"))
        display(Markdown(f"**Punkt:** $({sp.latex(x0_rational)},\\; {sp.latex(y0)})$, "
                         f"**Steigung:** $m = {sp.latex(m)}$"))
        display(Markdown(f"**Tangente:** $t(x) = {sp.latex(tangente)}$"))
        if normale is not None:
            display(Markdown(f"**Normale:** $n(x) = {sp.latex(normale)}$"))
        else:
            display(Markdown(f"**Normale:** $x = {sp.latex(x0_rational)}$ (senkrecht)"))

        # Plot
        f_num = sp.lambdify(_x, f_sym, 'numpy')
        t_num = sp.lambdify(_x, tangente, 'numpy')

        xs = np.linspace(float(x0_rational) - 3, float(x0_rational) + 3, 500)
        with np.errstate(all='ignore'):
            f_vals = np.array(f_num(xs), dtype=float)
            t_vals = np.array(t_num(xs), dtype=float)

        fig, ax = plt.subplots()
        ax.plot(xs, f_vals, 'b-', linewidth=2.5, label=rf'$f(x) = {sp.latex(f_sym)}$')
        ax.plot(xs, t_vals, 'r--', linewidth=1.8, label=rf'$t(x) = {sp.latex(tangente)}$')

        if normale is not None:
            n_num = sp.lambdify(_x, normale, 'numpy')
            with np.errstate(all='ignore'):
                n_vals = np.array(n_num(xs), dtype=float)
            ax.plot(xs, n_vals, 'g-.', linewidth=1.8, label=rf'$n(x) = {sp.latex(normale)}$')

        ax.plot(float(x0_rational), float(y0), 'ro', markersize=10, zorder=5)
        y_min = min(np.nanmin(f_vals), float(y0) - 5)
        y_max = max(np.nanmax(f_vals), float(y0) + 5)
        margin = (y_max - y_min) * 0.2
        ax.set_ylim(y_min - margin, y_max + margin)
        ax.axhline(y=0, color='k', linewidth=0.5)
        ax.axvline(x=0, color='k', linewidth=0.5)
        ax.grid(True, alpha=0.3)
        ax.legend(fontsize=11, loc='upper left', framealpha=0.9)
        ax.set_title('Tangente und Normale (symbolisch berechnet)', fontsize=14)
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Fehler: {e}")

interact(
    sympy_tangente,
    formel_str=Text(
        value='x**3 - 3*x', placeholder='z.B. x^3 - 3*x, e^x, x*exp(-x)',
        description='f(x) =', style={'description_width': '50px'},
        layout={'width': '470px'}, continuous_update=False,
    ),
    x0=FloatSlider(value=1.0, min=-3.0, max=3.0, step=0.5,
                   description='x\u2080:', style={'description_width': '30px'},
                   layout={'width': '400px'})
);
```

## Formelsammlung — Tangente, Normale & Funktionsscharen

### Tangente
$$\boxed{t(x) = f'(x_0) \cdot (x - x_0) + f(x_0)}$$

### Normale
$$\boxed{n(x) = -\frac{1}{f'(x_0)} \cdot (x - x_0) + f(x_0)} \qquad (f'(x_0) \neq 0)$$

### Berührbedingung (zwei Graphen berühren sich in $x_0$)
$$\boxed{f(x_0) = g(x_0) \quad \text{und} \quad f'(x_0) = g'(x_0)}$$

### Tangente durch externen Punkt P$(x_P | y_P)$
1. Tangente bei $x_0$ aufstellen: $t(x) = f'(x_0)(x - x_0) + f(x_0)$
2. P einsetzen: $y_P = f'(x_0)(x_P - x_0) + f(x_0)$
3. Nach $x_0$ auflösen → Berührstelle(n)

### Ortskurve bestimmen (3 Schritte)
1. **Extrempunkte** in Abhängigkeit vom Parameter: $x_E(t)$, $y_E(t)$
2. **Parameter isolieren:** aus $x_E(t)$ nach $t$ auflösen
3. **Einsetzen:** $t$ in $y_E(t)$ einsetzen → $y = g(x)$ ist die Ortskurve

### Waagrechte Tangente
$$f'(x_0) = 0 \quad \Rightarrow \quad t(x) = f(x_0) \quad \text{(konstant)}$$
