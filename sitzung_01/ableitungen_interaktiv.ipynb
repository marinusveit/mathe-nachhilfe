## 1. Was ist die Ableitung? — Tangente an einem Punkt

Verschiebe den Punkt entlang der Kurve und beobachte, wie sich die **Tangente** und ihre **Steigung** ändern.

```python
def tangente_explorer(x0=1.0):
    f = lambda x: x**3 - 3*x  # Funktion
    f_prime = lambda x: 3*x**2 - 3  # Ableitung
    
    x = np.linspace(-2.5, 2.5, 300)
    y = f(x)
    
    m = f_prime(x0)  # Steigung der Tangente
    y0 = f(x0)
    tangente = m * (x - x0) + y0  # Tangentengleichung
    
    fig, ax = plt.subplots()
    ax.plot(x, y, 'b-', linewidth=2, label=r'$f(x) = x^3 - 3x$')
    ax.plot(x, tangente, 'r--', linewidth=1.5,
            label=rf'Tangente ($m = {m:.2f}$)')
    ax.plot(x0, y0, 'ro', markersize=10, zorder=5)
    
    ax.set_xlim(-2.5, 2.5)
    ax.set_ylim(-6, 6)
    ax.axhline(y=0, color='k', linewidth=0.5)
    ax.axvline(x=0, color='k', linewidth=0.5)
    ax.grid(True, alpha=0.3)
    ax.legend(fontsize=12, loc='upper left', framealpha=0.9)
    ax.set_title(rf'Punkt bei $x_0 = {x0:.1f}$ $\;\rightarrow\;$ $f\'(x_0) = {m:.2f}$',
                 fontsize=14)
    plt.tight_layout()
    plt.show()

interact(
    tangente_explorer,
    x0=FloatSlider(value=1.0, min=-2.0, max=2.0, step=0.1, description='x₀:',
                   style={'description_width': '30px'}, layout={'width': '400px'})
);
```

**Beobachte:**
- Wo ist die Steigung **positiv**? Wo **negativ**? Wo **null**?
- Was passiert an den Stellen, wo f'(x₀) = 0 ist? → Das sind die **Extremstellen**!

## 2. f(x) und f'(x) im Vergleich

Hier siehst du eine Funktion und ihre Ableitung **gleichzeitig**. Wähle verschiedene Funktionen aus und entdecke den Zusammenhang.

```python
funktionen = {
    'x² − 2x': (lambda x: x**2 - 2*x, lambda x: 2*x - 2,
                r'$f(x) = x^2 - 2x$', r"$f'(x) = 2x - 2$"),
    'x³ − 3x': (lambda x: x**3 - 3*x, lambda x: 3*x**2 - 3,
                r'$f(x) = x^3 - 3x$', r"$f'(x) = 3x^2 - 3$"),
    'sin(x)': (lambda x: np.sin(x), lambda x: np.cos(x),
              r'$f(x) = \sin(x)$', r"$f'(x) = \cos(x)$"),
    'eˣ': (lambda x: np.exp(x), lambda x: np.exp(x),
           r'$f(x) = e^x$', r"$f'(x) = e^x$"),
    'ln(x)': (lambda x: np.log(np.maximum(x, 0.01)), lambda x: 1/np.maximum(x, 0.01),
             r'$f(x) = \ln(x)$', r"$f'(x) = 1/x$"),
}

def vergleich(wahl='x³ − 3x'):
    f, f_prime, f_label, fp_label = funktionen[wahl]
    
    if wahl == 'ln(x)':
        x = np.linspace(0.05, 5, 300)
    else:
        x = np.linspace(-3, 3, 300)
    
    fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(10, 7))
    
    # f(x)
    ax1.plot(x, f(x), 'b-', linewidth=2, label=f_label)
    ax1.axhline(y=0, color='k', linewidth=0.5)
    ax1.grid(True, alpha=0.3)
    ax1.legend(fontsize=14)
    ax1.set_title('Funktion', fontsize=13)
    
    # f'(x)
    ax2.plot(x, f_prime(x), 'r-', linewidth=2, label=fp_label)
    ax2.axhline(y=0, color='k', linewidth=0.5)
    ax2.grid(True, alpha=0.3)
    ax2.legend(fontsize=14)
    ax2.set_title('Ableitung', fontsize=13)
    
    # Nullstellen von f' markieren (= Extrema von f)
    fp_values = f_prime(x)
    for i in range(1, len(x)):
        if fp_values[i-1] * fp_values[i] < 0:  # Vorzeichenwechsel
            x_null = x[i]
            ax1.axvline(x=x_null, color='green', linestyle=':', alpha=0.7)
            ax2.axvline(x=x_null, color='green', linestyle=':', alpha=0.7)
            ax1.plot(x_null, f(x_null), 'go', markersize=10)
            ax2.plot(x_null, 0, 'go', markersize=10)
    
    plt.tight_layout()
    plt.show()

interact(
    vergleich,
    wahl=Dropdown(options=list(funktionen.keys()), value='x³ − 3x', description='Funktion:')
);
```

**Die grünen Punkte** zeigen: Wo f'(x) = 0 ist (Nullstelle der Ableitung), hat f(x) ein **Extremum**!

**Merkregeln:**
- f' > 0 → f steigt ↗
- f' < 0 → f fällt ↘
- f' = 0 → mögliches Extremum von f

## 3. Ableitungsregeln live testen

Hier kannst du mit **sympy** jede beliebige Funktion ableiten lassen und das Ergebnis direkt als Plot sehen.

```python
import sympy as sp
from ipywidgets import Text
from sympy.parsing.sympy_parser import parse_expr, standard_transformations, implicit_multiplication_application

_x = sp.Symbol('x')

_LOKALE_NAMEN = {
    'x': _x,
    'e': sp.E,
    'pi': sp.pi,
    'ln': sp.log,
    'log': sp.log,
    'exp': sp.exp,
    'sqrt': sp.sqrt,
    'sin': sp.sin,
    'cos': sp.cos,
    'tan': sp.tan,
}

_TRANSFORMATIONS = standard_transformations + (implicit_multiplication_application,)

def _plot_werte(expr, xs):
    with np.errstate(all='ignore'):
        werte = sp.lambdify(_x, expr, 'numpy')(xs)
    werte = np.asarray(werte, dtype=np.complex128)
    if werte.ndim == 0:
        werte = np.full(xs.shape, werte, dtype=np.complex128)
    reelle_werte = np.where(np.abs(werte.imag) < 1e-9, werte.real, np.nan)
    return np.where(np.isfinite(reelle_werte), reelle_werte, np.nan)

def _parse_formel(formel_str):
    formel_normiert = formel_str.strip().replace('^', '**')
    return parse_expr(
        formel_normiert,
        local_dict=_LOKALE_NAMEN,
        transformations=_TRANSFORMATIONS,
    )

def ableitung_zeigen(formel_str):
    """Gibt Funktion und Ableitung symbolisch aus und plottet beide."""
    try:
        f_sym = _parse_formel(formel_str)
        f_prime_sym = sp.diff(f_sym, _x)

        display(Markdown(f"**f(x) =** ${sp.latex(f_sym)}$"))
        display(Markdown(f"**f'(x) =** ${sp.latex(f_prime_sym)}$"))

        xs = np.linspace(-3, 3, 600)
        f_werte = _plot_werte(f_sym, xs)
        fp_werte = _plot_werte(f_prime_sym, xs)

        fig, ax = plt.subplots()
        ax.plot(xs, f_werte, 'b-', linewidth=2, label=r'$f(x)$')
        ax.plot(xs, fp_werte, 'r--', linewidth=2, label=r"$f'(x)$")
        ax.axhline(y=0, color='k', linewidth=0.5)
        ax.axvline(x=0, color='k', linewidth=0.5)
        ax.grid(True, alpha=0.3)
        ax.legend(fontsize=14, loc='upper left', framealpha=0.9)
        ax.set_xlim(xs[0], xs[-1])
        ax.set_ylim(-10, 10)
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Fehler: {e}")

interact(
    ableitung_zeigen,
    formel_str=Text(
        value='x**3 - 3*x',
        placeholder='z.B. x^4 - 2*x^2, e^x * x, ln(x)',
        description='f(x) =',
        style={'description_width': '50px'},
        layout={'width': '470px'},
        continuous_update=False,
    ),
);
```

### Probiere selbst! Ändere den String oben und drücke Enter, z.B.:

| Funktion | Eingabe |
|----------|--------|
| x⁴ − 2x² | `x^4 - 2*x^2` oder `x**4 - 2*x**2` |
| eˣ · x | `e^x * x` oder `exp(x) * x` |
| ln(x) | `ln(x)` oder `log(x)` |
| sin(πx²) | `sin(pi*x^2)` oder `sin(x**2)` |
| (2x+1)⁵ | `(2*x + 1)^5` |

## 4. Potenzregel — das Muster sehen

```python
def potenzregel_visual(n=2.0):
    """Zeigt f(x) = xⁿ und f'(x) = n·xⁿ⁻¹ für verschiedene n."""
    ist_ganzzahl = float(n).is_integer()
    n_wert = int(n) if ist_ganzzahl else n

    f_sym = _x**n_wert
    fp_sym = sp.diff(f_sym, _x)

    if ist_ganzzahl and n_wert < 0:
        xs_links = np.linspace(-2, -0.08, 300)
        xs_rechts = np.linspace(0.08, 2, 300)
        xs = np.concatenate([xs_links, xs_rechts])
    elif not ist_ganzzahl:
        xs = np.linspace(0.01, 3, 400)
    else:
        xs = np.linspace(-2, 2, 400)

    f_vals = _plot_werte(f_sym, xs)
    fp_vals = _plot_werte(fp_sym, xs)

    # LaTeX-Formatierung für Exponent
    n_str = f"{n:g}"
    nm1_str = f"{n-1:g}"

    fig, ax = plt.subplots()
    ax.plot(xs, f_vals, 'b-', linewidth=2.5,
            label=rf'$f(x) = x^{{{n_str}}}$')
    ax.plot(xs, fp_vals, 'r--', linewidth=2,
            label=rf"$f'(x) = {n_str} \cdot x^{{{nm1_str}}}$")
    ax.axhline(y=0, color='k', linewidth=0.5)
    ax.axvline(x=0, color='k', linewidth=0.5)
    ax.grid(True, alpha=0.3)
    ax.set_xlim(xs.min(), xs.max())
    ax.set_ylim(-5, 8)
    ax.legend(fontsize=14, loc='upper left', framealpha=0.9)
    ax.set_title(r"Potenzregel: $(x^n)' = n \cdot x^{n-1}$", fontsize=14)
    plt.tight_layout()
    plt.show()

interact(
    potenzregel_visual,
    n=FloatSlider(value=2.0, min=-2.0, max=5.0, step=0.5, description='n:',
                 style={'description_width': '30px'}, layout={'width': '400px'})
);
```

**Beobachte:**
- n = 2: Parabel → Gerade (Ableitung von x² = 2x)
- n = 3: Kubisch → Parabel
- n = 0.5: Wurzel → geht gegen ∞ bei x = 0
- n = -1: 1/x → -1/x²
