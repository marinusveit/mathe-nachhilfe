### Aufgabe A1: Ableitungen (4 BE)

```python
# A1a) f(x) = 3x^4 - 2x^2 + 5x - 1  →  Potenzregel
f_A1a = 3*x**4 - 2*x**2 + 5*x - 1
f_A1a_prime = sp.diff(f_A1a, x)
display(Markdown("**a)** Potenzregel:"))
display(Markdown(f"$f(x) = {sp.latex(f_A1a)}$"))
display(Markdown(f"$f'(x) = {sp.latex(f_A1a_prime)}$"))

# A1b) g(x) = (2x + 1)^5  →  Kettenregel
g_A1b = (2*x + 1)**5
g_A1b_prime = sp.diff(g_A1b, x)
display(Markdown("**b)** Kettenregel: äußere Ableitung $\\cdot$ innere Ableitung"))
display(Markdown(f"$g(x) = {sp.latex(g_A1b)}$"))
display(Markdown(f"$g'(x) = {sp.latex(g_A1b_prime)}$"))

# A1c) h(x) = x^2 * e^x  →  Produktregel
h_A1c = x**2 * sp.exp(x)
h_A1c_prime = sp.diff(h_A1c, x)
h_A1c_prime_simplified = sp.factor(h_A1c_prime)
display(Markdown("**c)** Produktregel: $(u \\cdot v)' = u' \\cdot v + u \\cdot v'$"))
display(Markdown(f"$h(x) = {sp.latex(h_A1c)}$"))
display(Markdown(f"$h'(x) = {sp.latex(h_A1c_prime)} = {sp.latex(h_A1c_prime_simplified)}$"))
```

### Aufgabe A2: Nullstellen und Extremum von $f(x) = x^3 - 3x$ (5 BE)

```python
# A2: f(x) = x^3 - 3x
f_A2 = x**3 - 3*x
f_A2_prime = sp.diff(f_A2, x)
f_A2_double_prime = sp.diff(f_A2, x, 2)

# a) Nullstellen
nst = sp.solve(f_A2, x)
display(Markdown("**a) Nullstellen:**"))
display(Markdown(f"$f(x) = {sp.latex(f_A2)} = x(x^2 - 3) = 0$"))
display(Markdown(f"$x_1 = 0, \\quad x_2 = \\sqrt{{3}}, \\quad x_3 = -\\sqrt{{3}}$"))

# b) Lokales Minimum bei x = 1
display(Markdown("**b) Nachweis: lokales Minimum bei $x = 1$**"))
display(Markdown(f"$f'(x) = {sp.latex(f_A2_prime)}$"))
display(Markdown(f"$f'(1) = {f_A2_prime.subs(x, 1)} = 0$ ✓ (notwendige Bedingung)"))
display(Markdown(f"$f''(x) = {sp.latex(f_A2_double_prime)}$"))
display(Markdown(f"$f''(1) = {f_A2_double_prime.subs(x, 1)} > 0$ → lokales **Minimum** ✓"))
display(Markdown(f"$f(1) = {f_A2.subs(x, 1)}$ → Tiefpunkt $T(1 \\mid -2)$"))

# Visualisierung
xs = np.linspace(-2.5, 2.5, 300)
f_np = lambda t: t**3 - 3*t

fig, ax = plt.subplots()
ax.plot(xs, f_np(xs), 'b-', linewidth=2, label=r'$f(x) = x^3 - 3x$')
ax.plot(1, -2, 'ro', markersize=10, zorder=5, label=r'Tiefpunkt $T(1 \mid -2)$')
ax.plot(-1, 2, 'go', markersize=10, zorder=5, label=r'Hochpunkt $H(-1 \mid 2)$')
for nx in [0, float(sp.sqrt(3)), -float(sp.sqrt(3))]:
    ax.plot(nx, 0, 'kx', markersize=10, markeredgewidth=2)
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.legend(fontsize=12, loc='upper left', framealpha=0.9)
ax.set_title('A2: Kurvendiskussion', fontsize=14)
plt.tight_layout()
plt.show()
```

### Aufgabe A4: Bestimmtes Integral (3 BE)

```python
# A4: Integral von 3x^2 - 4x + 1 auf [0, 2]
integrand = 3*x**2 - 4*x + 1
stammfunktion = sp.integrate(integrand, x)
ergebnis = sp.integrate(integrand, (x, 0, 2))

display(Markdown("**Lösung:**"))
display(Markdown(f"Stammfunktion: $F(x) = {sp.latex(stammfunktion)}$"))
display(Markdown(f"$\\displaystyle\\int_0^2 ({sp.latex(integrand)})\\, dx = "
                 f"\\left[{sp.latex(stammfunktion)}\\right]_0^2 = "
                 f"F(2) - F(0) = {stammfunktion.subs(x, 2)} - {stammfunktion.subs(x, 0)} = "
                 f"\\mathbf{{{ergebnis}}}$"))

# Visualisierung
xs = np.linspace(-0.5, 2.5, 300)
ys = 3*xs**2 - 4*xs + 1
xs_fill = np.linspace(0, 2, 300)
ys_fill = 3*xs_fill**2 - 4*xs_fill + 1

fig, ax = plt.subplots()
ax.plot(xs, ys, 'b-', linewidth=2, label=r'$f(x) = 3x^2 - 4x + 1$')
ax.fill_between(xs_fill, ys_fill, 0, alpha=0.3, color='blue',
                label=f'Integralwert = {float(ergebnis):.0f}')
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.legend(fontsize=12, loc='upper left', framealpha=0.9)
ax.set_title('A4: Bestimmtes Integral', fontsize=14)
plt.tight_layout()
plt.show()

display(Markdown("**Achtung:** Das Integral ist hier $= 2$, aber die Fläche ist *nicht* 2! "
                 "Der Graph taucht zwischen $x = 1/3$ und $x = 1$ unter die x-Achse. "
                 "Das Integral berücksichtigt das Vorzeichen — für den Flächeninhalt "
                 "müsste man den negativen Teil separat berechnen."))
```

### Aufgabe A5: Wahr oder Falsch (3 BE)

**a) Falsch.** $f'(x_0) = 0$ ist nur eine *notwendige* Bedingung. Gegenbeispiel: $f(x) = x^3$ hat $f'(0) = 0$, aber **kein** Extremum bei $x = 0$ (Sattelpunkt).

**b) Wahr.** $f''(x_0) = 0$ und $f'''(x_0) \neq 0$ ist die hinreichende Bedingung für einen Wendepunkt (Vorzeichenwechsel der Krümmung garantiert).

**c) Falsch.** Das bestimmte Integral $\int_a^b f(x)\, dx$ kann negativ werden, wenn $f(x) < 0$. Für den **Flächeninhalt** muss man $\int_a^b |f(x)|\, dx$ berechnen (Beträge!).

## Teil B — Mit Hilfsmitteln

### Aufgabe B1: $f(x) = (2x - x^2) \cdot e^x$ (30 BE)

```python
# ============================================================
# Teil B: f(x) = (2x - x^2) * e^x — Vollständige Kurvendiskussion
# ============================================================
f_B = (2*x - x**2) * sp.exp(x)
f_B_prime = sp.diff(f_B, x)
f_B_double_prime = sp.diff(f_B, x, 2)

display(Markdown("### a) Grenzverhalten und Nullstellen (4 BE)"))
display(Markdown(f"$f(x) = {sp.latex(f_B)}$"))

# Nullstellen
nst_B = sp.solve(f_B, x)
display(Markdown(f"**Nullstellen:** $f(x) = 0 \\Leftrightarrow (2x - x^2) \\cdot e^x = 0$"))
display(Markdown(f"Da $e^x > 0$ für alle $x$, gilt: $2x - x^2 = 0 \\Leftrightarrow x(2-x) = 0$"))
display(Markdown(f"$\\Rightarrow x_1 = {nst_B[0]}, \\quad x_2 = {nst_B[1]}$"))

# Grenzverhalten
lim_plus = sp.limit(f_B, x, sp.oo)
lim_minus = sp.limit(f_B, x, -sp.oo)
display(Markdown(f"**Grenzverhalten:**"))
display(Markdown(f"$\\lim_{{x \\to +\\infty}} f(x) = {sp.latex(lim_plus)}$ "
                 f"(da $-x^2$ dominiert und $e^x > 0$)"))
display(Markdown(f"$\\lim_{{x \\to -\\infty}} f(x) = {sp.latex(lim_minus)}$ "
                 f"(da $e^x \\to 0$ schneller als jedes Polynom wächst)"))
```

```python
# b) Extrempunkte (7 BE)
display(Markdown("### b) Extrempunkte (7 BE)"))

f_B_prime_factored = sp.factor(f_B_prime)
display(Markdown(f"$f'(x) = {sp.latex(sp.expand(f_B_prime))} = {sp.latex(f_B_prime_factored)}$"))

extremstellen = sp.solve(f_B_prime, x)
display(Markdown(f"$f'(x) = 0 \\Leftrightarrow (2 - x^2) \\cdot e^x = 0 "
                 f"\\Leftrightarrow x^2 = 2$"))
display(Markdown(f"$\\Rightarrow x_1 = -\\sqrt{{2}}, \\quad x_2 = \\sqrt{{2}}$"))

for xs_val in extremstellen:
    y_val = f_B.subs(x, xs_val)
    f2_val = f_B_double_prime.subs(x, xs_val)
    art = "Maximum (Hochpunkt)" if f2_val < 0 else "Minimum (Tiefpunkt)"
    display(Markdown(f"$x = {sp.latex(xs_val)}$: "
                     f"$f({sp.latex(xs_val)}) = {sp.latex(sp.simplify(y_val))}$, "
                     f"$f''({sp.latex(xs_val)}) = {sp.latex(sp.simplify(f2_val))}$ "
                     f"{'< 0' if f2_val < 0 else '> 0'} → **{art}**"))

display(Markdown(f"**Hochpunkt:** $H(\\sqrt{{2}} \\mid (2\\sqrt{{2}} - 2) \\cdot e^{{\\sqrt{{2}}}}) "
                 f"\\approx ({float(sp.sqrt(2)):.3f} \\mid {float(f_B.subs(x, sp.sqrt(2))):.3f})$"))
display(Markdown(f"**Tiefpunkt:** $T(-\\sqrt{{2}} \\mid (-2\\sqrt{{2}} - 2) \\cdot e^{{-\\sqrt{{2}}}}) "
                 f"\\approx ({-float(sp.sqrt(2)):.3f} \\mid {float(f_B.subs(x, -sp.sqrt(2))):.3f})$"))
```

```python
# c) Tangente bei x = -1 (5 BE)
display(Markdown("### c) Tangente im Punkt $P(-1 \\mid f(-1))$ (5 BE)"))

y0 = sp.simplify(f_B.subs(x, -1))
m = sp.simplify(f_B_prime.subs(x, -1))
display(Markdown(f"$f(-1) = {sp.latex(y0)}$, also $P(-1 \\mid {sp.latex(y0)})$"))
display(Markdown(f"$f'(-1) = {sp.latex(m)}$"))
tangente = sp.expand(m * (x + 1) + y0)
display(Markdown(f"**Tangentengleichung:** $t(x) = {sp.latex(tangente)}$"))
display(Markdown(f"**Nullstelle der Tangente:** $t(x) = 0 \\Leftrightarrow {sp.latex(tangente)} = 0 "
                 f"\\Leftrightarrow x = 2$"))
display(Markdown("Die Tangente schneidet die x-Achse also bei $x = 2$ und nicht im Berührpunkt."))
```

```python
# d) Flächenberechnung auf [0, 2] (6 BE)
display(Markdown("### d) Fläche auf $[0, 2]$ (6 BE)"))

# f >= 0 auf [0, 2] (Nullstellen bei 0 und 2, dazwischen positiv)
display(Markdown("Da $f(x) \\geq 0$ auf $[0, 2]$ (Nullstellen bei $x=0$ und $x=2$, "
                 "dazwischen positiv), gilt:"))

flaeche = sp.integrate(f_B, (x, 0, 2))
flaeche_simplified = sp.simplify(flaeche)

display(Markdown(f"$A = \\displaystyle\\int_0^2 (2x - x^2) \\cdot e^x \\, dx = "
                 f"{sp.latex(flaeche_simplified)} \\approx {float(flaeche_simplified):.4f}$"))

display(Markdown("**Rechenweg (partielle Integration):**"))
display(Markdown("Stammfunktion von $(2x - x^2) \\cdot e^x$ bestimmen "
                 "(2x partielle Integration oder CAS):"))
stammfunktion_B = sp.integrate(f_B, x)
display(Markdown(f"$F(x) = {sp.latex(sp.simplify(stammfunktion_B))}$"))
display(Markdown(f"$A = F(2) - F(0) = {sp.latex(sp.simplify(stammfunktion_B.subs(x, 2)))} "
                 f"- ({sp.latex(sp.simplify(stammfunktion_B.subs(x, 0)))}) "
                 f"= {sp.latex(flaeche_simplified)}$"))
```

```python
# e) Funktionenschar f_a(x) = (ax - x^2) * e^x (8 BE)
display(Markdown("### e) Funktionenschar (8 BE)"))

f_schar = (a*x - x**2) * sp.exp(x)
display(Markdown(f"$f_a(x) = {sp.latex(f_schar)}$"))

# Gemeinsame Nullstelle
display(Markdown("**Gemeinsame Nullstelle:**"))
display(Markdown(f"$f_a(x) = 0 \\Leftrightarrow (ax - x^2) \\cdot e^x = 0 "
                 f"\\Leftrightarrow x(a - x) = 0$"))
display(Markdown(f"$\\Rightarrow x_1 = 0$ (für alle $a$), $\\quad x_2 = a$"))
display(Markdown(f"Die gemeinsame Nullstelle ist $x = 0$."))

# Ortskurve der Hochpunkte
display(Markdown("**Ortskurve der Hochpunkte:**"))
f_schar_prime = sp.diff(f_schar, x)
f_schar_prime_factored = sp.factor(f_schar_prime)
display(Markdown(f"$f_a'(x) = {sp.latex(f_schar_prime_factored)}$"))

# f'(x) = 0 => (a + ax - x^2 - 2x) = 0 ... aber mit e^x > 0
extremstellen_schar = sp.solve(f_schar_prime, x)
display(Markdown(f"Kritische Stellen: $x = {sp.latex(extremstellen_schar[0])}$ "
                 f"und $x = {sp.latex(extremstellen_schar[1])}$"))

# Hochpunkt: das mit dem kleineren x-Wert? Nein, prüfen mit f''
# Für den Hochpunkt nehmen wir die Lösung, bei der f'' < 0
# Allgemein: x_H = (a-2+sqrt(a^2+4))/2 ... testen wir
display(Markdown("Wir setzen die Hochpunkt-Stelle in $f_a$ ein, um $y_H(a)$ zu erhalten. "
                 "Die Ortskurve ergibt sich durch Eliminieren von $a$:"))

# Einfacher Ansatz: x_H in f einsetzen
x_H = extremstellen_schar[1]  # die mit +sqrt
y_H = sp.simplify(f_schar.subs(x, x_H))
display(Markdown(f"$x_H = {sp.latex(x_H)}$"))
display(Markdown(f"$y_H = f_a(x_H) = {sp.latex(y_H)}$"))
display(Markdown("Die **Ortskurve** ist die Menge aller Hochpunkte $(x_H(a), y_H(a))$ "
                 "für $a > 0$ — parametrisch durch $a$ gegeben."))
```

# 4. Zusammenfassung aller Themen — Formelübersicht

## Ableitungsregeln

| Regel | Formel |
|-------|--------|
| Potenzregel | $(x^n)' = n \cdot x^{n-1}$ |
| Konstantenregel | $(c \cdot f)' = c \cdot f'$ |
| Summenregel | $(f + g)' = f' + g'$ |
| Produktregel | $(f \cdot g)' = f' \cdot g + f \cdot g'$ |
| Quotientenregel | $\left(\frac{f}{g}\right)' = \frac{f' \cdot g - f \cdot g'}{g^2}$ |
| Kettenregel | $(f(g(x)))' = f'(g(x)) \cdot g'(x)$ |

## Wichtige Ableitungen

| $f(x)$ | $f'(x)$ |
|---------|---------|
| $e^x$ | $e^x$ |
| $e^{kx}$ | $k \cdot e^{kx}$ |
| $\ln(x)$ | $\frac{1}{x}$ |
| $\sin(x)$ | $\cos(x)$ |
| $\cos(x)$ | $-\sin(x)$ |

## Kurvendiskussion — Schema

1. **Definitionsbereich** bestimmen
2. **Symmetrie** prüfen (gerade/ungerade)
3. **Nullstellen**: $f(x) = 0$
4. **Extrempunkte**: $f'(x) = 0$ (notwendig), $f''(x_0) \neq 0$ (hinreichend)
   - $f''(x_0) > 0$ → Minimum, $f''(x_0) < 0$ → Maximum
5. **Wendepunkte**: $f''(x) = 0$ (notwendig), $f'''(x_0) \neq 0$ (hinreichend)
6. **Monotonie**: $f'(x) > 0$ steigend, $f'(x) < 0$ fallend
7. **Verhalten** für $x \to \pm\infty$

## Integralrechnung

| Regel | Formel |
|-------|--------|
| Potenzregel | $\int x^n \, dx = \frac{x^{n+1}}{n+1} + C$ (für $n \neq -1$) |
| $\int \frac{1}{x}\, dx$ | $\ln|x| + C$ |
| $\int e^x \, dx$ | $e^x + C$ |
| $\int e^{kx} \, dx$ | $\frac{1}{k} \cdot e^{kx} + C$ |
| HDI | $\int_a^b f(x)\, dx = F(b) - F(a)$ |

## Flächenberechnung

- **Fläche zwischen Graph und x-Achse**: Nullstellen bestimmen, Integral aufteilen, Beträge addieren
- **Fläche zwischen zwei Graphen**: $A = \int_a^b |f(x) - g(x)|\, dx$

## Tangente

$$t(x) = f'(x_0) \cdot (x - x_0) + f(x_0)$$

## Funktionenschar

- Parameter $a$ als Konstante behandeln
- **Ortskurve**: $x_E(a)$ und $y_E(a)$ bestimmen, dann $a$ eliminieren
