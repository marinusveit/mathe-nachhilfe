```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from IPython.display import display, Markdown, HTML
from ipywidgets import interact, Checkbox, VBox, Layout, Accordion, Output
from contextlib import contextmanager

%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 5)
plt.rcParams['font.size'] = 13
plt.rcParams['text.usetex'] = False
plt.rcParams['mathtext.fontset'] = 'cm'  # Computer Modern (LaTeX-Look)

x = sp.Symbol('x')
a = sp.Symbol('a', positive=True)

# ─── Hilfsfunktionen für progressive Lösungsanzeige ───

@contextmanager
def loesung(titel="Lösung aufdecken"):
    out = Output()
    acc = Accordion(children=[out])
    acc.set_title(0, titel)
    acc.selected_index = None  # zugeklappt
    display(acc)
    with out:
        yield

def hinweis(text):
    display(HTML(
        f'<details style="background:#e8f4fd; border-left:4px solid #2196F3; '
        f'padding:10px; margin:8px 0; border-radius:4px;">'
        f'<summary style="cursor:pointer; font-weight:bold;">💡 Hinweis</summary>'
        f'<p style="margin-top:8px;">{text}</p></details>'
    ))

def typische_fehler(fehler_liste):
    items = ''.join(f'<li style="margin:4px 0;">{f}</li>' for f in fehler_liste)
    display(HTML(
        f'<details style="background:#fff3e0; border-left:4px solid #ff9800; '
        f'padding:10px; margin:8px 0; border-radius:4px;">'
        f'<summary style="cursor:pointer; font-weight:bold;">⚠️ Typische Fehler</summary>'
        f'<ul style="margin-top:8px;">{items}</ul></details>'
    ))
```

### Aufgabe A1: Ableitungen (4 BE) — *ca. 6 Min*

```python
hinweis("Welche Ableitungsregel brauchst du jeweils? a) Potenz, b) Kette, c) Produkt.")

with loesung("A1: Ableitungen — Lösung aufdecken"):
    # A1a) Potenzregel
    f_A1a = 3*x**4 - 2*x**2 + 5*x - 1
    f_A1a_prime = sp.diff(f_A1a, x)
    display(Markdown("**a)** Potenzregel:"))
    display(Markdown(f"$f(x) = {sp.latex(f_A1a)}$"))
    display(Markdown(f"$f'(x) = {sp.latex(f_A1a_prime)}$"))

    # A1b) Kettenregel
    g_A1b = (2*x + 1)**5
    g_A1b_prime = sp.diff(g_A1b, x)
    display(Markdown("**b)** Kettenregel: äußere Ableitung $\\cdot$ innere Ableitung"))
    display(Markdown(f"$g(x) = {sp.latex(g_A1b)}$"))
    display(Markdown(f"$g'(x) = {sp.latex(g_A1b_prime)}$"))

    # A1c) Produktregel
    h_A1c = x**2 * sp.exp(x)
    h_A1c_prime = sp.diff(h_A1c, x)
    h_A1c_prime_simplified = sp.factor(h_A1c_prime)
    display(Markdown("**c)** Produktregel: $(u \\cdot v)' = u' \\cdot v + u \\cdot v'$"))
    display(Markdown(f"$h(x) = {sp.latex(h_A1c)}$"))
    display(Markdown(f"$h'(x) = {sp.latex(h_A1c_prime)} = {sp.latex(h_A1c_prime_simplified)}$"))

    typische_fehler([
        "<b>Kettenregel vergessen:</b> Bei b) nur <code>5·(2x+1)⁴</code> statt <code>10·(2x+1)⁴</code> — die innere Ableitung (= 2) fehlt!",
        "<b>Produktregel nicht angewendet:</b> Bei c) nur <code>2x · eˣ</code> geschrieben statt <code>(2x + x²) · eˣ</code>.",
        "<b>Vorzeichen bei a):</b> Aus <code>-2x²</code> wird <code>-4x</code>, nicht <code>+4x</code>.",
    ])
```

### Aufgabe A2: Nullstellen und Extremum von $f(x) = x^3 - 3x$ (5 BE) — *ca. 8 Min*

```python
hinweis("a) Klammere x aus. b) Notwendige Bedingung: f'(1)=0? Hinreichende Bedingung: f''(1)>0?")

with loesung("A2: Nullstellen & Extremum — Lösung aufdecken"):
    f_A2 = x**3 - 3*x
    f_A2_prime = sp.diff(f_A2, x)
    f_A2_double_prime = sp.diff(f_A2, x, 2)

    # a) Nullstellen
    nst = sp.solve(f_A2, x)
    display(Markdown("**a) Nullstellen:**"))
    display(Markdown(f"$f(x) = {sp.latex(f_A2)} = x(x^2 - 3) = 0$"))
    display(Markdown(f"$x_1 = 0, \\quad x_2 = \\sqrt{{3}}, \\quad x_3 = -\\sqrt{{3}}$"))

    # b) Lokales Minimum bei x = 1
    display(Markdown("**b) Nachweis: lokales Minimum bei $x = 1$**"))
    display(Markdown(f"$f'(x) = {sp.latex(f_A2_prime)}$"))
    display(Markdown(f"$f'(1) = {f_A2_prime.subs(x, 1)} = 0$ ✓ (notwendige Bedingung)"))
    display(Markdown(f"$f''(x) = {sp.latex(f_A2_double_prime)}$"))
    display(Markdown(f"$f''(1) = {f_A2_double_prime.subs(x, 1)} > 0$ → lokales **Minimum** ✓"))
    display(Markdown(f"$f(1) = {f_A2.subs(x, 1)}$ → Tiefpunkt $T(1 \\mid -2)$"))

    # Visualisierung
    xs = np.linspace(-2.5, 2.5, 300)
    f_np = lambda t: t**3 - 3*t

    fig, ax = plt.subplots()
    ax.plot(xs, f_np(xs), 'b-', linewidth=2, label=r'$f(x) = x^3 - 3x$')
    ax.plot(1, -2, 'ro', markersize=10, zorder=5, label=r'Tiefpunkt $T(1 \mid -2)$')
    ax.plot(-1, 2, 'go', markersize=10, zorder=5, label=r'Hochpunkt $H(-1 \mid 2)$')
    for nx in [0, float(sp.sqrt(3)), -float(sp.sqrt(3))]:
        ax.plot(nx, 0, 'kx', markersize=10, markeredgewidth=2)
    ax.axhline(y=0, color='k', linewidth=0.5)
    ax.axvline(x=0, color='k', linewidth=0.5)
    ax.grid(True, alpha=0.3)
    ax.legend(fontsize=12, loc='upper left', framealpha=0.9)
    ax.set_title('A2: Kurvendiskussion', fontsize=14)
    plt.tight_layout()
    plt.show()

    typische_fehler([
        "<b>Nur f'(1)=0 gezeigt:</b> Das ist nur die notwendige Bedingung — ohne f''(1)>0 ist der Nachweis unvollständig!",
        "<b>Nullstellen vergessen auszuklammern:</b> f(x) = x(x²-3) ist der Schlüssel.",
        "<b>Verwechslung:</b> f''(1) = 6·1 = 6 > 0 heißt Minimum, nicht Maximum (Eselsbrücke: positiv = 'fröhliches Gesicht' = Tal).",
    ])
```

### Aufgabe A3: Graphen zuordnen (5 BE) — *ca. 8 Min*

```python
# A3: Graph der Funktion f (Aufgabenstellung)
# Wir verwenden f(x) = x^3 - 6x^2 + 9x - 1 (wird der Schülerin NICHT verraten)
f_A3_np = lambda t: t**3 - 6*t**2 + 9*t - 1

xs = np.linspace(-0.5, 4.5, 500)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(xs, f_A3_np(xs), 'b-', linewidth=2.5, label=r'Graph von $f$')
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.set_xlabel('$x$', fontsize=14)
ax.set_ylabel('$y$', fontsize=14)
ax.set_title('Aufgabe A3: Welcher Graph gehört zu $f\'$?', fontsize=14)
ax.legend(fontsize=13, loc='upper right', framealpha=0.9)
plt.tight_layout()
plt.show()
```

```python
hinweis("Wo hat f Extrema? Dort muss f' Nullstellen haben. Fällt f zwischen den Extrema? Dann ist f' dort negativ.")

with loesung("A3: Graphen zuordnen — Lösung aufdecken"):
    display(Markdown("**Lösung:** Graph I ist korrekt."))
    display(Markdown(
        "**Begründung:** Der Graph von $f$ hat:\n"
        "- ein **Maximum** bei ca. $x \\approx 1$ → dort muss $f'$ eine **Nullstelle mit VZW von + nach −** haben\n"
        "- ein **Minimum** bei ca. $x \\approx 3$ → dort muss $f'$ eine **Nullstelle mit VZW von − nach +** haben\n"
        "- dazwischen fällt $f$ → $f' < 0$ in diesem Bereich\n\n"
        "**Graph I** passt: nach oben geöffnete Parabel, Nullstellen bei $x = 1$ und $x = 3$, dazwischen **negativ**.\n"
        "**Graph II** scheidet aus: dazwischen positiv → $f$ würde steigen (Widerspruch).\n"
        "**Graph III** scheidet aus: Gerade → $f$ wäre quadratisch (passt nicht zum Verlauf)."
    ))

    # Alle drei Kandidaten plotten
    f_A3 = x**3 - 6*x**2 + 9*x - 1
    f_A3_prime = sp.diff(f_A3, x)  # 3x^2 - 12x + 9 = 3(x-1)(x-3)

    xs = np.linspace(-0.5, 4.5, 500)
    f_prime_np = lambda t: 3*t**2 - 12*t + 9       # Graph I (korrekt)
    graph_II_np = lambda t: -(3*t**2 - 12*t + 9)   # Graph II (invertiert)
    graph_III_np = lambda t: -2*t                    # Graph III (Gerade)

    fig, axes = plt.subplots(1, 3, figsize=(15, 4))

    for ax_i, (fn, titel, farbe, korrekt) in zip(axes, [
        (f_prime_np, 'Graph I', '#2ecc71', True),
        (graph_II_np, 'Graph II', '#e74c3c', False),
        (graph_III_np, 'Graph III', '#e74c3c', False),
    ]):
        ax_i.plot(xs, fn(xs), linewidth=2, color=farbe)
        ax_i.axhline(y=0, color='k', linewidth=0.5)
        ax_i.axvline(x=0, color='k', linewidth=0.5)
        ax_i.grid(True, alpha=0.3)
        ax_i.set_title(f'{titel} {"✓" if korrekt else "✗"}', fontsize=13,
                       color='green' if korrekt else 'red')
        ax_i.set_ylim(-15, 15)

    plt.suptitle("Kandidaten für $f'(x)$", fontsize=14)
    plt.tight_layout()
    plt.show()

    typische_fehler([
        "<b>Verwechslung steigend/fallend:</b> f steigt → f' > 0 (nicht umgekehrt!).",
        "<b>Graph II gewählt:</b> Positive Werte zwischen den Nullstellen würden bedeuten, dass f dort steigt — aber f fällt!",
        "<b>Nur auf Nullstellen geschaut:</b> Die Lage der Nullstellen allein reicht nicht — das Vorzeichen dazwischen entscheidet.",
    ])
```

### Aufgabe A4: Bestimmtes Integral (3 BE) — *ca. 4 Min*

```python
hinweis("Stammfunktion bilden, dann Grenzen einsetzen: F(2) - F(0).")

with loesung("A4: Bestimmtes Integral — Lösung aufdecken"):
    integrand = 3*x**2 - 4*x + 1
    stammfunktion = sp.integrate(integrand, x)
    ergebnis = sp.integrate(integrand, (x, 0, 2))

    display(Markdown("**Lösung:**"))
    display(Markdown(f"Stammfunktion: $F(x) = {sp.latex(stammfunktion)}$"))
    display(Markdown(f"$\\displaystyle\\int_0^2 ({sp.latex(integrand)})\\, dx = "
                     f"\\left[{sp.latex(stammfunktion)}\\right]_0^2 = "
                     f"F(2) - F(0) = {stammfunktion.subs(x, 2)} - {stammfunktion.subs(x, 0)} = "
                     f"\\mathbf{{{ergebnis}}}$"))

    # Visualisierung
    xs = np.linspace(-0.5, 2.5, 300)
    ys = 3*xs**2 - 4*xs + 1
    xs_fill = np.linspace(0, 2, 300)
    ys_fill = 3*xs_fill**2 - 4*xs_fill + 1

    fig, ax = plt.subplots()
    ax.plot(xs, ys, 'b-', linewidth=2, label=r'$f(x) = 3x^2 - 4x + 1$')
    ax.fill_between(xs_fill, ys_fill, 0, where=(ys_fill >= 0), alpha=0.3, color='blue',
                    label='positive Fläche')
    ax.fill_between(xs_fill, ys_fill, 0, where=(ys_fill < 0), alpha=0.3, color='red',
                    label='negative Fläche')
    ax.axhline(y=0, color='k', linewidth=0.5)
    ax.axvline(x=0, color='k', linewidth=0.5)
    ax.grid(True, alpha=0.3)
    ax.legend(fontsize=12, loc='upper left', framealpha=0.9)
    ax.set_title(f'A4: Bestimmtes Integral = {float(ergebnis):.0f} (aber Fläche ≠ Integral!)', fontsize=14)
    plt.tight_layout()
    plt.show()

    display(Markdown("**Achtung:** Das Integral ist hier $= 2$, aber die **Fläche** ist *nicht* 2! "
                     "Der Graph taucht zwischen $x = 1/3$ und $x = 1$ unter die x-Achse. "
                     "Das Integral berücksichtigt das Vorzeichen — für den **Flächeninhalt** "
                     "müsste man $\\int_0^{1/3} f\\, dx - \\int_{1/3}^{1} f\\, dx + \\int_1^2 f\\, dx$ berechnen."))

    typische_fehler([
        "<b>Stammfunktion falsch gebildet:</b> Potenzregel rückwärts: x<sup>n</sup> → x<sup>n+1</sup>/(n+1). Aus 3x² wird x³ (nicht 6x³!).",
        "<b>F(0) vergessen:</b> Hier ist F(0) = 0, aber das ist nicht immer so — immer F(b) - F(a) rechnen.",
        "<b>Integral = Fläche gesetzt:</b> Wenn f teilweise negativ ist, liefert das Integral NICHT den Flächeninhalt.",
    ])
```

### Aufgabe A5: Wahr oder Falsch (3 BE) — *ca. 4 Min*

```python
hinweis("Denke an Gegenbeispiele! Bei a) und c) hilft jeweils ein konkretes Beispiel.")

with loesung("A5: Wahr oder Falsch — Lösung aufdecken"):
    display(Markdown(
        "**a) Falsch.** $f'(x_0) = 0$ ist nur eine *notwendige* Bedingung.\n"
        "Gegenbeispiel: $f(x) = x^3$ hat $f'(0) = 0$, aber **kein** Extremum bei $x = 0$ (Sattelpunkt).\n\n"
        "**b) Wahr.** $f''(x_0) = 0$ und $f'''(x_0) \\\\neq 0$ ist die hinreichende Bedingung für einen Wendepunkt (Vorzeichenwechsel der Krümmung garantiert).\n\n"
        "**c) Falsch.** Das bestimmte Integral $\\\\int_a^b f(x)\\\\, dx$ kann negativ werden, wenn $f(x) < 0$.\n"
        "Für den **Flächeninhalt** muss man $\\\\int_a^b |f(x)|\\\\, dx$ berechnen (Beträge!)."
    ))

    typische_fehler([
        "<b>a) als 'wahr' beantwortet:</b> Klassiker! f'(x₀) = 0 ist NOTWENDIG, aber NICHT HINREICHEND. Immer an Sattelpunkte denken.",
        "<b>c) als 'wahr' beantwortet:</b> Integral kann negativ sein, Fläche nie. Merke: Fläche = Integral mit Beträgen.",
        "<b>Keine Begründung:</b> 'Wahr' oder 'Falsch' allein gibt keine volle Punktzahl — immer kurz begründen!",
    ])
```

### Aufgabe B1: $f(x) = (2x - x^2) \cdot e^x$ (30 BE)

```python
hinweis("Nullstellen: Setze f(x)=0. Beachte: e<sup>x</sup> > 0 für alle x! Grenzwert: e<sup>x</sup> dominiert jedes Polynom.")

with loesung("B1a: Grenzverhalten & Nullstellen (4 BE, ca. 7 Min) — Lösung aufdecken"):
    f_B = (2*x - x**2) * sp.exp(x)
    f_B_prime = sp.diff(f_B, x)
    f_B_double_prime = sp.diff(f_B, x, 2)

    display(Markdown(f"$f(x) = {sp.latex(f_B)}$"))

    # Nullstellen
    nst_B = sp.solve(f_B, x)
    display(Markdown(f"**Nullstellen:** $f(x) = 0 \\Leftrightarrow (2x - x^2) \\cdot e^x = 0$"))
    display(Markdown(f"Da $e^x > 0$ für alle $x$, gilt: $2x - x^2 = 0 \\Leftrightarrow x(2-x) = 0$"))
    display(Markdown(f"$\\Rightarrow x_1 = {nst_B[0]}, \\quad x_2 = {nst_B[1]}$"))

    # Grenzverhalten
    lim_plus = sp.limit(f_B, x, sp.oo)
    lim_minus = sp.limit(f_B, x, -sp.oo)
    display(Markdown(f"**Grenzverhalten:**"))
    display(Markdown(f"$\\lim_{{x \\to +\\infty}} f(x) = {sp.latex(lim_plus)}$ "
                     f"(da $-x^2 \\cdot e^x \\to -\\infty$ dominiert)"))
    display(Markdown(f"$\\lim_{{x \\to -\\infty}} f(x) = {sp.latex(lim_minus)}$ "
                     f"(da $e^x \\to 0$ schneller als jedes Polynom wächst)"))

    typische_fehler([
        "<b>e<sup>x</sup> = 0 als Lösung:</b> Die e-Funktion wird NIE null! Nur der Polynomteil kann null werden.",
        "<b>Grenzwert für x → +∞ falsch:</b> Trotz e<sup>x</sup> → ∞ geht f(x) → -∞, weil der Faktor -x² dominiert.",
        "<b>Klammer vergessen aufzulösen:</b> 2x - x² = x(2 - x), nicht x(2 + x).",
    ])
```

```python
hinweis("Ableitung mit Produktregel bilden, dann f'(x)=0 lösen. Für die Art: f''(x₀) auswerten.")

with loesung("B1b: Extrempunkte (7 BE, ca. 12 Min) — Lösung aufdecken"):
    f_B_prime_factored = sp.factor(f_B_prime)
    display(Markdown(f"$f'(x) = {sp.latex(sp.expand(f_B_prime))} = {sp.latex(f_B_prime_factored)}$"))

    extremstellen = sp.solve(f_B_prime, x)
    display(Markdown(f"$f'(x) = 0 \\Leftrightarrow (2 - x^2) \\cdot e^x = 0 "
                     f"\\Leftrightarrow x^2 = 2$"))
    display(Markdown(f"$\\Rightarrow x_1 = -\\sqrt{{2}}, \\quad x_2 = \\sqrt{{2}}$"))

    for xs_val in extremstellen:
        y_val = f_B.subs(x, xs_val)
        f2_val = f_B_double_prime.subs(x, xs_val)
        art = "Maximum (Hochpunkt)" if f2_val < 0 else "Minimum (Tiefpunkt)"
        display(Markdown(f"$x = {sp.latex(xs_val)}$: "
                         f"$f({sp.latex(xs_val)}) = {sp.latex(sp.simplify(y_val))}$, "
                         f"$f''({sp.latex(xs_val)}) = {sp.latex(sp.simplify(f2_val))}$ "
                         f"{'< 0' if f2_val < 0 else '> 0'} → **{art}**"))

    display(Markdown(f"**Hochpunkt:** $H(\\sqrt{{2}} \\mid (2\\sqrt{{2}} - 2) \\cdot e^{{\\sqrt{{2}}}}) "
                     f"\\approx ({float(sp.sqrt(2)):.3f} \\mid {float(f_B.subs(x, sp.sqrt(2))):.3f})$"))
    display(Markdown(f"**Tiefpunkt:** $T(-\\sqrt{{2}} \\mid (-2\\sqrt{{2}} - 2) \\cdot e^{{-\\sqrt{{2}}}}) "
                     f"\\approx ({-float(sp.sqrt(2)):.3f} \\mid {float(f_B.subs(x, -sp.sqrt(2))):.3f})$"))

    typische_fehler([
        "<b>Produktregel bei f'(x) vergessen:</b> f(x) = u·v mit u = (2x-x²) und v = e<sup>x</sup>. Beide Teile ableiten!",
        "<b>e<sup>x</sup> = 0 gelöst:</b> Wieder: e<sup>x</sup> > 0, also nur 2-x² = 0 lösen.",
        "<b>Hochpunkt und Tiefpunkt verwechselt:</b> f'' < 0 → Hochpunkt (Linkskurve), f'' > 0 → Tiefpunkt (Rechtskurve).",
    ])
```

```python
hinweis("Tangentenformel: t(x) = f'(x₀) · (x - x₀) + f(x₀). Hier ist x₀ = -1.")

with loesung("B1c: Tangente in P(-1|f(-1)) (5 BE, ca. 8 Min) — Lösung aufdecken"):
    display(Markdown("### c) Tangente im Punkt $P(-1 \\mid f(-1))$ (5 BE)"))

    x0 = -1
    y0 = sp.simplify(f_B.subs(x, x0))
    m = sp.simplify(f_B_prime.subs(x, x0))

    display(Markdown(f"**Schritt 1:** Funktionswert berechnen"))
    display(Markdown(f"$f(-1) = (2 \\cdot (-1) - (-1)^2) \\cdot e^{{-1}} = (-2 - 1) \\cdot e^{{-1}} = {sp.latex(y0)}$"))

    display(Markdown(f"**Schritt 2:** Steigung berechnen"))
    display(Markdown(f"$f'(x) = (2 - x^2) \\cdot e^x$"))
    display(Markdown(f"$f'(-1) = (2 - 1) \\cdot e^{{-1}} = {sp.latex(m)}$"))

    display(Markdown(f"**Schritt 3:** Tangentengleichung aufstellen"))
    tangente = sp.expand(m * (x - x0) + y0)
    display(Markdown(f"$t(x) = f'(-1) \\cdot (x - (-1)) + f(-1) = "
                     f"{sp.latex(m)} \\cdot (x + 1) + \\left({sp.latex(y0)}\\right)$"))
    display(Markdown(f"$t(x) = \\frac{{x + 1}}{{e}} - \\frac{{3}}{{e}} = \\frac{{x - 2}}{{e}}$"))

    display(Markdown(f"**Schritt 4:** Nullstelle der Tangente"))
    tangente_nst = sp.solve(tangente, x)
    display(Markdown(f"$t(x) = 0 \\Leftrightarrow \\frac{{x - 2}}{{e}} = 0 \\Leftrightarrow x = {tangente_nst[0]}$"))
    display(Markdown(f"Die Tangente schneidet die x-Achse bei $x = 2$ — das ist auch eine Nullstelle von $f$!"))

    typische_fehler([
        "<b>Formel falsch angewendet:</b> t(x) = f'(x₀)·(x - x₀) + f(x₀), NICHT t(x) = f'(x₀)·x + f(x₀).",
        "<b>Vorzeichen bei x₀ = -1:</b> (x - (-1)) = (x + 1), nicht (x - 1)!",
        "<b>f'(x) statt f'(x₀) eingesetzt:</b> Die Steigung muss eine Zahl sein, keine Funktion von x.",
    ])
```

```python
hinweis("f ≥ 0 auf [0,2], also direkte Integration. Tipp: Schreibe (2x-x²)·eˣ als Summe (2x·eˣ - x²·eˣ) und integriere jeweils partiell.")

with loesung("B1d: Fläche auf [0, 2] (6 BE, ca. 10 Min) — Lösung aufdecken"):
    display(Markdown("### d) Fläche auf $[0, 2]$ (6 BE)"))

    display(Markdown("Da $f(x) \\geq 0$ auf $[0, 2]$ (Nullstellen bei $x=0$ und $x=2$, "
                     "dazwischen positiv), gilt:"))
    display(Markdown("$A = \\displaystyle\\int_0^2 (2x - x^2) \\cdot e^x \\, dx$"))

    display(Markdown("---"))
    display(Markdown("**Ausführlicher Rechenweg — Partielle Integration Schritt für Schritt:**"))

    display(Markdown(
        "Wir zerlegen das Integral:\n\n"
        "$\\\\displaystyle\\\\int (2x - x^2) \\\\cdot e^x \\\\, dx = \\\\int 2x \\\\cdot e^x \\\\, dx - \\\\int x^2 \\\\cdot e^x \\\\, dx$\n\n"
        "**Integral 1:** $\\\\int 2x \\\\cdot e^x \\\\, dx$ (1x partiell integrieren)\n\n"
        "| | $u$ | $v'$ |\n"
        "|---|---|---|\n"
        "| | $2x$ | $e^x$ |\n"
        "| ableiten/integrieren | $2$ | $e^x$ |\n\n"
        "$= 2x \\\\cdot e^x - \\\\int 2 \\\\cdot e^x \\\\, dx = 2x \\\\cdot e^x - 2e^x$\n\n"
        "**Integral 2:** $\\\\int x^2 \\\\cdot e^x \\\\, dx$ (2x partiell integrieren)\n\n"
        "*1. Runde:*\n\n"
        "| | $u$ | $v'$ |\n"
        "|---|---|---|\n"
        "| | $x^2$ | $e^x$ |\n"
        "| ableiten/integrieren | $2x$ | $e^x$ |\n\n"
        "$= x^2 \\\\cdot e^x - \\\\int 2x \\\\cdot e^x \\\\, dx$\n\n"
        "*2. Runde:* $\\\\int 2x \\\\cdot e^x \\\\, dx$ kennen wir schon von oben!\n\n"
        "$= x^2 \\\\cdot e^x - (2x \\\\cdot e^x - 2e^x) = x^2 \\\\cdot e^x - 2x \\\\cdot e^x + 2e^x$\n\n"
        "**Zusammensetzen:**\n"
        "$\\\\int (2x - x^2) \\\\cdot e^x \\\\, dx = (2x \\\\cdot e^x - 2e^x) - (x^2 \\\\cdot e^x - 2x \\\\cdot e^x + 2e^x)$\n\n"
        "$= 2x \\\\cdot e^x - 2e^x - x^2 \\\\cdot e^x + 2x \\\\cdot e^x - 2e^x$\n\n"
        "$= (-x^2 + 4x - 4) \\\\cdot e^x$\n\n"
        "Also: $F(x) = (-x^2 + 4x - 4) \\\\cdot e^x$"
    ))

    # Numerische Verifikation
    F_hand = (-x**2 + 4*x - 4) * sp.exp(x)
    flaeche = sp.simplify(F_hand.subs(x, 2) - F_hand.subs(x, 0))

    display(Markdown(f"**Einsetzen der Grenzen:**"))
    display(Markdown(f"$F(2) = (-4 + 8 - 4) \\cdot e^2 = 0$"))
    display(Markdown(f"$F(0) = (0 + 0 - 4) \\cdot e^0 = -4$"))
    display(Markdown(f"$A = F(2) - F(0) = 0 - (-4) = \\mathbf{{4}}$"))

    # Sympy-Check
    flaeche_sympy = sp.integrate(f_B, (x, 0, 2))
    display(Markdown(f"**Kontrolle mit CAS:** $A = {sp.latex(sp.simplify(flaeche_sympy))} "
                     f"\\approx {float(sp.simplify(flaeche_sympy)):.4f}$"))

    # Hinweis: Sympy und Handrechnung können leicht abweichen in der Stammfunktion
    # (unterschiedliche Konstanten), aber das Ergebnis ist gleich

    typische_fehler([
        "<b>Vergessen dass f ≥ 0:</b> Ohne Vorzeichenprüfung weiß man nicht, ob man Beträge braucht.",
        "<b>Partielle Integration falsch angesetzt:</b> u = Polynom, v' = e<sup>x</sup> — NICHT umgekehrt!",
        "<b>Zweite Runde vergessen:</b> Bei x²·e<sup>x</sup> muss man ZWEIMAL partiell integrieren.",
        "<b>Vorzeichen beim Zusammensetzen:</b> Das Minus vor dem zweiten Integral wirkt auf die GANZE Klammer.",
    ])
```

```python
hinweis("Gemeinsame Nullstelle: Welcher x-Wert ist unabhängig von a? Ortskurve: f'<sub>a</sub>(x)=0 lösen → x<sub>H</sub>(a), dann a eliminieren.")

with loesung("B1e: Funktionenschar & Ortskurve (8 BE, ca. 13 Min) — Lösung aufdecken"):
    display(Markdown("### e) Funktionenschar (8 BE)"))

    f_schar = (a*x - x**2) * sp.exp(x)
    display(Markdown(f"$f_a(x) = {sp.latex(f_schar)}$"))

    # --- Gemeinsame Nullstelle ---
    display(Markdown("**Gemeinsame Nullstelle:**"))
    display(Markdown(f"$f_a(x) = 0 \\Leftrightarrow (ax - x^2) \\cdot e^x = 0 "
                     f"\\Leftrightarrow x(a - x) = 0$"))
    display(Markdown(f"$\\Rightarrow x_1 = 0$ (für alle $a > 0$), $\\quad x_2 = a$ (hängt von $a$ ab)"))
    display(Markdown(f"Die **gemeinsame** Nullstelle aller Scharfunktionen ist $x = 0$. ✓"))

    # --- Ortskurve der Hochpunkte ---
    display(Markdown("---"))
    display(Markdown("**Ortskurve der Hochpunkte — Schritt für Schritt:**"))

    f_schar_prime = sp.diff(f_schar, x)
    f_schar_prime_factored = sp.factor(f_schar_prime)
    display(Markdown(f"**Schritt 1:** Ableitung bilden"))
    display(Markdown(f"$f_a'(x) = {sp.latex(f_schar_prime_factored)}$"))

    display(Markdown(f"**Schritt 2:** Extremstellen bestimmen ($f_a'(x) = 0$)"))
    display(Markdown(f"Da $e^x > 0$: $\\quad a + ax - 2x - x^2 = 0$"))
    display(Markdown(f"Umformen: $x^2 + (2-a)x - a = 0$"))

    extremstellen_schar = sp.solve(f_schar_prime, x)
    display(Markdown(f"Mitternachtsformel: $x_{{1,2}} = \\frac{{a-2 \\pm \\sqrt{{(2-a)^2 + 4a}}}}{{2}} "
                     f"= \\frac{{a-2 \\pm \\sqrt{{a^2+4}}}}{{2}}$"))

    display(Markdown(f"**Schritt 3:** Hochpunkt identifizieren"))
    display(Markdown(f"Da $\\sqrt{{a^2+4}} > |a-2|$ für alle $a > 0$, ist die Lösung mit '$+$' immer positiv:"))
    x_H = extremstellen_schar[1]
    display(Markdown(f"$x_H = {sp.latex(x_H)}$"))

    display(Markdown(f"**Schritt 4:** $y_H$ berechnen — $x_H$ in $f_a$ einsetzen"))
    y_H = sp.simplify(f_schar.subs(x, x_H))
    display(Markdown(f"$y_H = f_a(x_H) = {sp.latex(y_H)}$"))

    display(Markdown(f"**Schritt 5:** Parameter $a$ eliminieren"))
    display(Markdown(f"Aus $x_H = \\frac{{a-2+\\sqrt{{a^2+4}}}}{{2}}$ folgt:"))
    display(Markdown(f"$2x_H = a - 2 + \\sqrt{{a^2+4}}$"))
    display(Markdown(f"$2x_H - a + 2 = \\sqrt{{a^2+4}}$"))
    display(Markdown(f"Quadrieren: $(2x_H - a + 2)^2 = a^2 + 4$"))
    display(Markdown(f"$4x_H^2 + a^2 + 4 - 4ax_H + 8x_H - 4a = a^2 + 4$"))
    display(Markdown(f"$4x_H^2 - 4ax_H + 8x_H - 4a = 0$"))
    display(Markdown(f"$4x_H^2 + 8x_H = 4a(x_H + 1)$"))
    display(Markdown(f"$a = \\frac{{x_H^2 + 2x_H}}{{x_H + 1}} = \\frac{{x_H(x_H + 2)}}{{x_H + 1}}$"))
    display(Markdown(f"Diesen Ausdruck für $a$ kann man in $y_H(a)$ einsetzen, um die Ortskurve "
                     f"$y = g(x)$ als explizite Funktion zu erhalten."))
    display(Markdown(f"Die **Ortskurve** ist die Menge aller Hochpunkte $(x_H(a), y_H(a))$ "
                     f"für $a > 0$ — parametrisch durch $a$ gegeben."))

    typische_fehler([
        "<b>'Gemeinsame Nullstelle x = a':</b> x = a ist NICHT gemeinsam, weil es von a abhängt! Nur x = 0 gilt für alle a.",
        "<b>Ortskurve: a nicht eliminiert:</b> Die Ortskurve muss als y = g(x) angegeben werden, nicht parametrisch. (In der Praxis akzeptieren viele Korrektoren auch parametrisch.)",
        "<b>Falsches Extremum gewählt:</b> Es gibt zwei kritische Stellen — prüfe mit f'' welche der Hochpunkt ist.",
    ])
```

# 4. Zusammenfassung aller Themen — Formelübersicht

## Ableitungsregeln

| Regel | Formel |
|-------|--------|
| Potenzregel | $(x^n)' = n \cdot x^{n-1}$ |
| Konstantenregel | $(c \cdot f)' = c \cdot f'$ |
| Summenregel | $(f + g)' = f' + g'$ |
| Produktregel | $(f \cdot g)' = f' \cdot g + f \cdot g'$ |
| Quotientenregel | $\left(\frac{f}{g}\right)' = \frac{f' \cdot g - f \cdot g'}{g^2}$ |
| Kettenregel | $(f(g(x)))' = f'(g(x)) \cdot g'(x)$ |

## Wichtige Ableitungen

| $f(x)$ | $f'(x)$ |
|---------|---------|
| $e^x$ | $e^x$ |
| $e^{kx}$ | $k \cdot e^{kx}$ |
| $\ln(x)$ | $\frac{1}{x}$ |
| $\sin(x)$ | $\cos(x)$ |
| $\cos(x)$ | $-\sin(x)$ |

## Kurvendiskussion — Schema

1. **Definitionsbereich** bestimmen
2. **Symmetrie** prüfen (gerade/ungerade)
3. **Nullstellen**: $f(x) = 0$
4. **Extrempunkte**: $f'(x) = 0$ (notwendig), $f''(x_0) \neq 0$ (hinreichend)
   - $f''(x_0) > 0$ → Minimum, $f''(x_0) < 0$ → Maximum
5. **Wendepunkte**: $f''(x) = 0$ (notwendig), $f'''(x_0) \neq 0$ (hinreichend)
6. **Monotonie**: $f'(x) > 0$ steigend, $f'(x) < 0$ fallend
7. **Verhalten** für $x \to \pm\infty$

## Integralrechnung

| Regel | Formel |
|-------|--------|
| Potenzregel | $\int x^n \, dx = \frac{x^{n+1}}{n+1} + C$ (für $n \neq -1$) |
| $\int \frac{1}{x}\, dx$ | $\ln|x| + C$ |
| $\int e^x \, dx$ | $e^x + C$ |
| $\int e^{kx} \, dx$ | $\frac{1}{k} \cdot e^{kx} + C$ |
| HDI | $\int_a^b f(x)\, dx = F(b) - F(a)$ |

## Flächenberechnung

- **Fläche zwischen Graph und x-Achse**: Nullstellen bestimmen, Integral aufteilen, Beträge addieren
- **Fläche zwischen zwei Graphen**: $A = \int_a^b |f(x) - g(x)|\, dx$

## Tangente

$$t(x) = f'(x_0) \cdot (x - x_0) + f(x_0)$$

## Funktionenschar

- Parameter $a$ als Konstante behandeln
- **Ortskurve**: $x_E(a)$ und $y_E(a)$ bestimmen, dann $a$ eliminieren
